```python
import pandas as pd
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side

# Donn√©es du tableau r√©capitulatif global
data = {
    "Module": [
        "1. √âtudiant",
        "2. Professeur",
        "3. Sous-admin",
        "4. Cours / Module",
        "5. Examen & Note",
        "6. Calendrier",
        "7. Auth & R√¥les",
        "8. √âl√©ments communs",
        "9. Tests globaux",
        "10. Page d‚Äôaccueil",
        "11. UI/UX Design",
        "12. D√©ploiement & Docs"
    ],
    "Dur√©e (jours)": [7, 5, 3.5, 6, 6, 5, 6, 3, 2.5, 4, 6, 4],
    "Co√ªt (MAD)": [6300, 4500, 3150, 5400, 5400, 4500, 5400, 2700, 2250, 3600, 5400, 3600]
}

# Cr√©ation du DataFrame
df = pd.DataFrame(data)

# Ajout de la ligne de total
total_row = pd.DataFrame({
    "Module": ["üü¶ TOTAL G√âN√âRAL"],
    "Dur√©e (jours)": [sum(df["Dur√©e (jours)"])],
    "Co√ªt (MAD)": [sum(df["Co√ªt (MAD)"])]
})
df_final = pd.concat([df, total_row], ignore_index=True)

# Cr√©ation du fichier Excel avec openpyxl
wb = Workbook()
ws = wb.active
ws.title = "Devis Technique"

# Ajout des en-t√™tes
headers = list(df_final.columns)
ws.append(headers)

# Styles
header_fill = PatternFill(start_color="4472C4", end_color="4472C4", fill_type="solid")
header_font = Font(color="FFFFFF", bold=True)
center_alignment = Alignment(horizontal="center", vertical="center")
thin_border = Border(left=Side(style='thin'), right=Side(style='thin'),
                     top=Side(style='thin'), bottom=Side(style='thin'))

# Style d‚Äôen-t√™te
for cell in ws[1]:
    cell.fill = header_fill
    cell.font = header_font
    cell.alignment = center_alignment
    cell.border = thin_border

# Donn√©es
for row in df_final.itertuples(index=False):
    ws.append(row)

# Mise en forme des cellules
for row in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=1, max_col=3):
    for cell in row:
        cell.border = thin_border
        cell.alignment = Alignment(vertical="center")

# Ligne de total
total_row_idx = ws.max_row
for cell in ws[total_row_idx]:
    cell.font = Font(bold=True, color="000000")
    cell.fill = PatternFill(start_color="D9E1F2", end_color="D9E1F2", fill_type="solid")

# Ajustement automatique des colonnes
for column in ws.columns:
    max_length = 0
    column_letter = column[0].column_letter
    for cell in column:
        if cell.value:
            max_length = max(max_length, len(str(cell.value)))
    ws.column_dimensions[column_letter].width = max_length + 2

# Sauvegarde du fichier
wb.save("Devis_Technique.xlsx")
print("‚úÖ Fichier Excel styl√© g√©n√©r√© : Devis_Technique.xlsx")
```
